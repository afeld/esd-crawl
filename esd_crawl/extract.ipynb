```python
from glob import glob
from itertools import islice
from esd_crawl import pipelines
```

```python
def pages_with_tables(pdfs_path):
    for path in glob(pdfs_path):
        yield from pipelines.pages_with_tables(path)
```

```python
PDFS_PATH = "../downloads/**/*.pdf"
NUM_TO_DISPLAY = 5

pages = pages_with_tables(PDFS_PATH)
for page in islice(pages, 0, NUM_TO_DISPLAY):
    print(f"Page number: {page.page_number}")

    img = page.to_image()
    finder = img.debug_tablefinder()
    display(finder)
```
