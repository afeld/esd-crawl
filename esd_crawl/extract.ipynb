```python
from concurrent.futures import process
from glob import glob
from itertools import islice
import pdfplumber
```

```python
PDFS_PATH = "../downloads/**/*.pdf"

def pages_with_tables():
    for path in glob(PDFS_PATH):
        with pdfplumber.open(path) as pdf:
            for page in pdf.pages:
                tables = page.find_tables()
                if len(tables) > 0:
                    yield page
```

```python

NUM_TO_DISPLAY = 5

pages = pages_with_tables()
for page in islice(pages, 0, NUM_TO_DISPLAY):
    img = page.to_image()
    finder = img.debug_tablefinder()
    display(finder)
```
